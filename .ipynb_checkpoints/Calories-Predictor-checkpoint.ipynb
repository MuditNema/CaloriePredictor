# Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

 # Data Collection 

```python
calories = pd.read_csv('./calories.csv')
exercise = pd.read_csv('./exercise.csv')
```

# Combining the two data frames

```python
calories_info = pd.concat([exercise,calories['Calories']],axis=1)
calories_info.head()
```

##### Checking for missing values

```python
calories_info.isnull().sum()
#We have 0 missing values
```

# Analysing the data-set

```python
sns.set()
sns.countplot(calories_info['Gender'])
```

```python
sns.distplot(calories_info['Age'])
```

# Finding the correlation in the dataset
## Types of correlation :

### 1. Positive correlation
### 2. Negative correlation

```python
correlation = calories_info.corr()
```

```python
# Constructing a heat map to understand the correlation

plt.figure(figsize=(10,10))
sns.heatmap(correlation , cbar=True,square=True, fmt='.1f', annot=True, annot_kws = {'size':8},cmap='Blues')
```

# Updating the data-types to fit for the ML Algorithm

```python
# Since Gender values are not in numbers (they are in string)
# We will convert 
# male -> 1
# female -> 0

calories_info.replace({'Gender':{'male':1,'female':0}},inplace=True)
# calories_info.head()
```

# Separating features and target/prediction class

```python
X = calories_info.drop(columns=['User_ID','Calories'],axis=1)
Y = calories_info['Calories']
```

# Splitting the data into training data and test data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X, Y, test_size=0.3, random_state=8)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

# Model training using XGBoost Regressor

```python
XGB_model = XGBRegressor()

#Training the model with X_train

XGB_model.fit(X_train, Y_train)
```

# Evaluating our model

```python
# test_data_prediction = XGB_model.predict(X_test)

#In the app , we'll be using a different test data set
# Gender	Age	Height	Weight	Duration	Heart_Rate	Body_Temp
X = np.array([[1,28,174,81,25,150,39]])
test_data_prediction = XGB_model.predict(X)
print(test_data_prediction)
```

# Saving our model for our application using pickle

```python
import pickle
data = {
    "model":XGB_model
}
with open('saved_steps.pkl','wb') as file:
    pickle.dump(data,file)
```

```python
with open('saved_steps.pkl','rb') as file:
    data = pickle.load(file)
    
print(data["model"].predict(X))  
```
